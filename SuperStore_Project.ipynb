```python
data['Ship Date'] = np.where( 
    (data.shipping_timeframe <0) & (data['Ship Mode'] == 'First Class'),
                             data['Order Date'] + timedelta(days=3), 
                             data['Ship Date']
                             )
```

```python
data['Ship Date'] = np.where( 
    (data.shipping_timeframe <0) & (data['Ship Mode'] == 'Same Day'),
                             data['Order Date'], 
                             data['Ship Date']
                             )
```

```python
data['Ship Date'] = np.where( 
    (data.shipping_timeframe <0) & (data['Ship Mode'] == 'Second Class'),
                             data['Order Date']+ timedelta(days=4), 
                             data['Ship Date']
                             )
```

```python
data['Ship Date'] = np.where( 
    (data.shipping_timeframe <0) & (data['Ship Mode'] == 'Standard Class'),
                             data['Order Date']+ timedelta(days=6), 
                             data['Ship Date']
                             )
```

```python
ship_mode_dict = {'First Class':3,
                  'Same Day':0,
                  'Second Class':4,
                  'Standard Class':6}
```

```python
ship_mode_dict.items()
```

```python
for key,value in ship_mode_dict.items():
  data['Ship Date'] = np.where( 
    (data.shipping_timeframe <0) & (data['Ship Mode'] == key),
                             data['Order Date'] + timedelta(days=value), 
                             data['Ship Date']
                             )

```
